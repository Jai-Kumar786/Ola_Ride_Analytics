```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load the Enhanced Dataset ---
# We start with the feature-rich dataset created on Day 3.
try:
    df = pd.read_csv('../data/ola_data_enhanced.csv')
    df['booking_timestamp'] = pd.to_datetime(df['booking_timestamp'])
    print("✅ Enhanced dataset loaded successfully.")
except FileNotFoundError:
    print("❌ Error: 'ola_data_enhanced.csv' not found. Make sure you ran the Day 3 script.")
    exit()

# --- 2. Bivariate Analysis ---
# Exploring the relationships between two variables to find patterns and correlations.
sns.set_style("whitegrid")
print("\n--- Starting Bivariate Analysis ---")

# -- Plot 1: Booking Value vs. Ride Distance --
# We expect a positive correlation: longer rides should cost more.
plt.figure(figsize=(12, 7))
sns.scatterplot(x='ride_distance', y='booking_value', data=df[df['booking_status']=='Success'], alpha=0.5)
plt.title('Booking Value vs. Ride Distance for Successful Rides', fontsize=16)
plt.xlabel('Ride Distance (km)', fontsize=12)
plt.ylabel('Booking Value', fontsize=12)
plt.show()

# -- Plot 2: Driver Ratings by Vehicle Type --
# Do certain vehicle types consistently get better ratings?
plt.figure(figsize=(12, 7))
sns.boxplot(x='vehicle_type', y='driver_ratings', data=df, order=df['vehicle_type'].value_counts().index, palette='coolwarm')
plt.title('Driver Ratings Distribution by Vehicle Type', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Driver Rating', fontsize=12)
plt.show()

# -- Plot 3: Cancellations by Hour of Day --
# This directly investigates if cancellations have a time-based pattern.
plt.figure(figsize=(12, 7))
# We create a new DataFrame to isolate cancellations for this plot
cancellations_df = df[df['booking_status'].str.contains('Canceled')]
sns.countplot(x='hour_of_day', data=cancellations_df, palette='viridis')
plt.title('Total Cancellations by Hour of the Day', fontsize=16)
plt.xlabel('Hour of Day (24-hour format)', fontsize=12)
plt.ylabel('Number of Cancellations', fontsize=12)
plt.show()


# --- Goal for Excellence: Correlation Heatmap ---
print("\n--- Generating Correlation Heatmap ---")
plt.figure(figsize=(14, 10))

# Select only the numeric columns for the correlation matrix
numeric_cols = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
corr_matrix = numeric_cols.corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Numeric Variables', fontsize=16)
plt.show()


# --- 3. Formulate Hypotheses ---
# Based on our entire EDA process, we can now state our key questions formally.
# These will be the guiding questions for our deep-dive analysis in SQL and Power BI.
print("\n--- Formulating Key Hypotheses ---")
hypotheses = [
    "Hypothesis 1: Rides are most frequently cancelled during the evening peak hours (18:00-21:00), likely due to a mismatch between high demand and driver availability.",
    "Hypothesis 2: The average customer rating is significantly lower for rides with a longer ride distance, suggesting a potential decline in service quality or comfort on extended trips.",
    "Hypothesis 3: 'Prime Sedan' and 'Prime Plus', being premium categories, receive significantly higher average driver and customer ratings compared to economy options like 'Auto' and 'Bike'."
]

for h in hypotheses:
    print(f"✅ {h}")
```
